```python
def hello_outlier(df=None, column=None, weight=1.5):
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)

    IQR = quantile_75 - quantile_25
    IQR_weight = IQR * weight

    lowest = quantile_25 - IQR_weight
    highest = quantile_75 + IQR_weight

    outlier_idx = df[column][(df[column] < lowest) |
                             (df[column] > highest)].index
    return outlier_idx


def bye_outlier(df=None, column=None, weight=1.5):
    outlier_idx = hello_outlier(df=df, column=column, weight=1.5)
    df.drop(outlier_idx, axis=0, inplace=True)
```

# 우리 모델이 나올 수 있는 경우의 수  (About Outliers)
  
### 1. 아웃라이어 기본설정 (Weight = 1.5) 
####     1-1 10,000개가 넘는 데이터 그대로 삭제 전부 삭제 (too much list)
####     1-2 '하루최저습도'와 '하루최고습도' 칼럼 해당 row만 삭제
####     1-3 XG / LGBM / RF 의 Feature Importance와 Permutation Importance, Correlation 3개가 동시에 내놓는 중요 Feature에 대해서만 아웃라이어 처리

### 2. 아웃라이어 완화설정 (Wight > 1.5) 
####     2-1 아웃라이어가 최대 1,000개가 넘지 않게 조정 (기준 : 임의 설정) 後 1,000개가 넘는 row그대로 삭제

####     3-1 아웃라이어[기본설정]이면서 target Data에서 '보통' 범주에 속하는 row만 삭제
####     3-2 아웃라이어[완화설정]이면서 target Data에서 '보통' 범주에 속하는 row만 삭제

```python
def outlier_line(df=None, column=None, weight=2) : 
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)

    IQR = quantile_75 - quantile_25
    IQR_weight = IQR*weight

    lowest = quantile_25 - IQR_weight
    highest = quantile_75 + IQR_weight
    
    return lowest, highest
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
from matplotlib import pyplot

from matplotlib import font_manager, rc
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import StratifiedKFold, KFold, cross_val_score, TimeSeriesSplit

warnings.filterwarnings(action='ignore')
%matplotlib inline

font_path = 'C:\\Users\\JeongSeongYun\\AppData\\Local\\Microsoft\\Windows\\Fonts\\NanumBarunGothic.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

numeric = [
    '이산화질소농도(ppm)', '오존농도(ppm)', '이산화탄소농도(ppm)', '아황산가스(ppm)', '주변공사횟수',
    '인허가 환경사업수', '인허가 오염사업수', '도로비율(%)', '가로수 노선수', '가로수 수량', '정화식물 가로수',
    '하루평균강수량', '하루최저강수량', '하루최저습도', '하루최고습도', '하루최저풍속', '하루최고풍속'
]

will_be_numeric = ['측정지역', '풍향', '하늘상태']

col = [
    '이산화질소농도(ppm)', '오존농도(ppm)', '이산화탄소농도(ppm)', '아황산가스(ppm)', '주변공사횟수',
    '인허가 환경사업수', '인허가 오염사업수', '도로비율(%)', '가로수 노선수', '가로수 수량', '정화식물 가로수',
    '하루평균강수량', '하루최저강수량', '하루최고강수량', '하루평균습도', '하루최저습도', '하루최고습도', '일최고기온',
    '일최저기온', '하루최저풍속', '하루최고풍속', '미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)'
]


def hello_outlier(df=None, column=None, weight=1.5):
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)

    IQR = quantile_75 - quantile_25
    IQR_weight = IQR * weight

    lowest = quantile_25 - IQR_weight
    highest = quantile_75 + IQR_weight

    outlier_idx = df[column][(df[column] < lowest) |
                             (df[column] > highest)].index
    return outlier_idx


def bye_outlier(df=None, column=None, weight=1.5):
    outlier_idx = hello_outlier(df=df, column=column, weight=1.5)
    df.drop(outlier_idx, axis=0, inplace=True)


def outlier_line(df=None, column=None, weight=2):
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)

    IQR = quantile_75 - quantile_25
    IQR_weight = IQR * weight

    lowest = quantile_25 - IQR_weight
    highest = quantile_75 + IQR_weight

    return lowest, highest
```
